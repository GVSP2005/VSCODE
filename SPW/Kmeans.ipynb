```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
points=np.random.rand(100,2)
kmeans=KMeans(n_clusters=8,random_state=234)
kmeans.fit(points)
```

```python
labels=kmeans.labels_
cluster_points=[[] for i in range(8)]
for i,label in enumerate(labels):
    cluster_points[label].append(points[i])
x_min=points[:,0].min()-1
x_max=points[:,0].max()+1
y_min=points[:,1].min()-1
y_max=points[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,0.001),np.arange(y_min,y_max,0.001))
z=kmeans.predict(np.c_[xx.ravel(),yy.ravel()])
z=z.reshape(xx.shape)
plt.contour(xx,yy,z,alpha=0.56)
for i in range(8):
    plt.scatter(np.array(cluster_points[i])[:,0],np.array(cluster_points[i])[:,1],label='data{}'.format(i))
plt.legend()
```
